```python
import cv2 
from copy import deepcopy
import matplotlib.pyplot as plt

%matplotlib inline
plt.style.use('ggplot')
```

```python
yo_mero = cv2.imread('yomero.jpeg')
```

```python
yo_mero.max(), yo_mero.shape
```

```python
yo_en_rgb = cv2.cvtColor(yo_mero, cv2.COLOR_BGR2RGB)

plt.figure()
plt.imshow(yo_en_rgb)
plt.axis('off')
plt.title('Foto tomada en nochebuena')
plt.show()
```

```python
yo_resized= cv2.resize(yo_en_rgb, (480, 800))

fig, (ax) = plt.subplots(nrows=1, figsize=(5, 15))
ax.imshow(yo_resized)
ax.set_title('Foto tomada en nochebuena')
plt.show()
```

```python
yo_cropped = yo_resized[260:320, 100:280]

fig, axes = plt.subplots(nrows=1, ncols=4, figsize=(20, 5))

axes[0].imshow(yo_resized)
axes[0].set_title('Yo listo para comer mi peso en pavo')

axes[1].imshow(yo_cropped)
axes[1].set_title('ojos')

cuatrojos = deepcopy(yo_resized)
ojos = cv2.resize(yo_cropped, (180, 90))
cuatrojos[150:240, 90:270] = ojos

axes[2].imshow(ojos)
axes[2].set_title('resized ojos')

axes[3].imshow(cuatrojos)
axes[3].set_title('Qué he hecho...')

plt.tight_layout()
plt.show()
```
